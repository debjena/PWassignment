**STATISTICS BASIC**

```python
# ğŸ“˜ PW Skills - Statistics Assignment

# âœ… 1. What is statistics, and why is it important?
# Statistics is the science of collecting, analyzing, interpreting, and presenting data.
# It is important because it helps in making informed decisions using data.

# âœ… 2. What are the two main types of statistics?
# 1. Descriptive Statistics
# 2. Inferential Statistics

# âœ… 3. What are descriptive statistics?
# These summarize data through numbers, charts, or graphs.
# Examples: mean, median, mode, range, variance, standard deviation.

# âœ… 4. What is inferential statistics?
# It uses a sample to make predictions or inferences about a population.
# Example: hypothesis testing, confidence intervals.

# âœ… 5. What is sampling in statistics?
# Sampling is selecting a subset from a population to estimate characteristics of the whole.

# âœ… 6. What are the different types of sampling methods?
# 1. Simple Random
# 2. Stratified
# 3. Cluster
# 4. Systematic
# 5. Convenience

# âœ… 7. What is the difference between random and non-random sampling?
# Random: Equal chance to every member, less bias.
# Non-Random: Based on judgment or convenience, more bias.

# âœ… 8. Define and give examples of qualitative and quantitative data
# Qualitative: Non-numeric (e.g., gender, color)
# Quantitative: Numeric (e.g., age, height)

# âœ… 9. What are the different types of data in statistics?
# 1. Nominal
# 2. Ordinal
# 3. Interval
# 4. Ratio

# âœ… 10. Explain nominal, ordinal, interval, and ratio levels of measurement
# Nominal: Categories (e.g., gender)
# Ordinal: Ordered categories (e.g., rankings)
# Interval: Numeric, no true zero (e.g., temperature)
# Ratio: Numeric with true zero (e.g., weight, height)

# âœ… 11. What is the measure of central tendency?
# It describes the center of a dataset: mean, median, and mode.

# âœ… 12. Define mean, median, and mode
import numpy as np
from scipy import stats

data = [1, 2, 2, 3, 4, 5]
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data, keepdims=True).mode[0])

# âœ… 13. What is the significance of the measure of central tendency?
# It summarizes a large dataset with a single representative value.

# âœ… 14. What is variance, and how is it calculated?
data2 = [2, 4, 6, 8, 10]
print("Variance:", np.var(data2))

# âœ… 15. What is standard deviation, and why is it important?
print("Standard Deviation:", np.std(data2))
# It shows how spread out the data is from the mean.

# âœ… 16. Define and explain the term range in statistics
print("Range:", max(data2) - min(data2))

# âœ… 17. What is the difference between variance and standard deviation?
# Variance: average squared deviations.
# Standard deviation: square root of variance, same unit as data.

# âœ… 18. What is skewness in a dataset?
from scipy.stats import skew
print("Skewness:", skew(data2))

# âœ… 19. What does it mean if a dataset is positively or negatively skewed?
# Positively skewed: Tail on the right, mean > median.
# Negatively skewed: Tail on the left, mean < median.

# âœ… 20. Define and explain kurtosis
from scipy.stats import kurtosis
print("Kurtosis:", kurtosis(data2))
# High kurtosis = more outliers, Low = fewer outliers.

# âœ… 21. What is the purpose of covariance?
x = [1, 2, 3]
y = [4, 5, 6]
print("Covariance:", np.cov(x, y)[0, 1])

# âœ… 22. What does correlation measure in statistics?
print("Correlation:", np.corrcoef(x, y)[0, 1])
# Correlation shows direction & strength of relationship.

# âœ… 23. What is the difference between covariance and correlation?
# Covariance: Direction only, not scaled.
# Correlation: Direction and strength, scaled (-1 to +1).

# âœ… 24. What are some real-world applications of statistics?
# - Business analytics
# - Medical trials
# - Machine learning
# - Sports performance
# - Government policymaking
```

**Practical**

```python


import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# âœ… 1. Calculate Mean, Median, Mode
data = [12, 15, 14, 10, 18, 20, 15]
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data, keepdims=True).mode[0])

# âœ… 2. Variance and Standard Deviation
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))

# âœ… 3. Classify dataset: Nominal, Ordinal, Interval, Ratio
dataset = {
    'Nominal': ['Red', 'Blue', 'Green'],
    'Ordinal': ['Low', 'Medium', 'High'],
    'Interval': [20, 30, 40],   # Celsius temperature
    'Ratio': [10, 20, 30]       # Age or weight
}
print("Classified dataset types:", dataset)

# âœ… 4. Random and Stratified Sampling
df = pd.DataFrame({
    'Gender': ['M', 'F']*5,
    'Marks': [65, 70, 80, 90, 85, 75, 60, 95, 55, 50]
})

# Random Sampling
random_sample = df.sample(n=4)
print("\nRandom Sample:\n", random_sample)

# Stratified Sampling
from sklearn.model_selection import train_test_split
strat_sample = df.groupby('Gender', group_keys=False).apply(lambda x: x.sample(2))
print("\nStratified Sample:\n", strat_sample)

# âœ… 5. Function to calculate Range
def calculate_range(data):
    return max(data) - min(data)
print("Range of dataset:", calculate_range(data))

# âœ… 6. Plot Histogram to visualize Skewness
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram of Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# âœ… 7. Skewness & Kurtosis using Python
print("Skewness:", stats.skew(data))
print("Kurtosis:", stats.kurtosis(data))

# âœ… 8. Positive & Negative Skewness Demo
pos_skew = [1, 2, 3, 4, 5, 100]
neg_skew = [100, 95, 90, 85, 80, 1]

sns.histplot(pos_skew, kde=True)
plt.title("Positive Skewness")
plt.show()

sns.histplot(neg_skew, kde=True)
plt.title("Negative Skewness")
plt.show()

# âœ… 9. Covariance between two datasets
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
cov_matrix = np.cov(x, y)
print("Covariance:", cov_matrix[0][1])

# âœ… 10. Correlation Coefficient
corr = np.corrcoef(x, y)
print("Correlation Coefficient:", corr[0][1])

# âœ… 11. Scatter Plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()

# âœ… 12. Compare Random vs Systematic Sampling
# Systematic Sampling
def systematic_sampling(df, step):
    return df.iloc[::step]

sys_sample = systematic_sampling(df, 2)
print("\nSystematic Sample:\n", sys_sample)

# âœ… 13. Grouped data mean, median, mode
grouped_data = {
    "Class Interval": ["0-10", "10-20", "20-30", "30-40"],
    "Frequency": [5, 8, 15, 7]
}
df_grouped = pd.DataFrame(grouped_data)
print("\nGrouped Data:\n", df_grouped)

# Mean â‰ˆ Î£(f*x)/Î£f (using midpoints)
midpoints = [5, 15, 25, 35]
frequencies = df_grouped["Frequency"]
mean_grouped = np.average(midpoints, weights=frequencies)
print("Grouped Mean:", mean_grouped)

# âœ… 14. Simulate data and compute central tendency & dispersion
sim_data = np.random.randint(10, 100, 20)
print("Simulated Data:", sim_data)
print("Mean:", np.mean(sim_data))
print("Median:", np.median(sim_data))
print("Std Dev:", np.std(sim_data))

# âœ… 15. Use NumPy/Pandas to describe a dataset
df_stats = pd.DataFrame({'Marks': sim_data})
print(df_stats.describe())

# âœ… 16. Boxplot for Outliers
sns.boxplot(x=sim_data)
plt.title("Boxplot of Simulated Data")
plt.show()

# âœ… 17. Interquartile Range (IQR)
Q1 = np.percentile(sim_data, 25)
Q3 = np.percentile(sim_data, 75)
IQR = Q3 - Q1
print("Interquartile Range (IQR):", IQR)

# âœ… 18. Z-score normalization
scaler = StandardScaler()
z_scores = scaler.fit_transform(np.array(sim_data).reshape(-1, 1))
print("Z-score normalized data:\n", z_scores.flatten())

# âœ… 19. Compare standard deviations of 2 datasets
data_a = np.random.randint(10, 50, 10)
data_b = np.random.randint(40, 90, 10)
print("Std Dev A:", np.std(data_a))
print("Std Dev B:", np.std(data_b))

# âœ… 20. Heatmap for Covariance
df_cov = pd.DataFrame({'x': x, 'y': y})
sns.heatmap(df_cov.cov(), annot=True)
plt.title("Covariance Heatmap")
plt.show()

# âœ… 21. Correlation Matrix using seaborn
sns.heatmap(df_cov.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# âœ… 22. Compute both Variance & Standard Deviation
var = np.var(data)
std = np.std(data)
print("Variance:", var, "\nStandard Deviation:", std)

# âœ… 23. Visualize skewness & kurtosis
sns.histplot(data, kde=True)
plt.title("Distribution with Skewness and Kurtosis")
plt.show()

# âœ… 24. Pearson & Spearman Correlation
print("Pearson:", stats.pearsonr(x, y)[0])
print("Spearman:", stats.spearmanr(x, y)[0])
```
