```python
# ðŸ“˜ PYTHON FUNCTION THEORY â€“ With Examples

# 1. Difference between a function and a method in Python
# âž¤ A function is a block of code that performs a task.
# âž¤ A method is a function associated with an object (like a string or list).

def greet():
    return "Hello"

print(greet())  # Function

text = "hello"
print(text.upper())  # Method applied on string object

# 2. Function arguments and parameters
# âž¤ Parameters are variables listed in function definition.
# âž¤ Arguments are values passed during the function call.

def add(x, y):  # x and y are parameters
    return x + y

print(add(5, 3))  # 5 and 3 are arguments

# 3. Ways to define and call a function
# âž¤ Using def, lambda, and calling with positional or keyword arguments

def greet(name):
    print(f"Hello, {name}!")

greet("Alice")                 # Positional
greet(name="Bob")             # Keyword

# 4. Purpose of `return` statement
# âž¤ It ends the function and sends a result back to the caller.

def multiply(a, b):
    return a * b

result = multiply(3, 4)
print(result)

# 5. Iterators vs Iterables
# âž¤ Iterable: Can be looped over (like list, tuple)
# âž¤ Iterator: Object with __next__() method that gives next value

nums = [1, 2, 3]
it = iter(nums)
print(next(it))  # 1
print(next(it))  # 2

# 6. Generators
# âž¤ A generator is a function with `yield` instead of `return`.
# âž¤ It saves state and produces values one at a time.

def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):
    print(num)

# 7. Advantages of Generators
# âž¤ Save memory (donâ€™t store entire sequence)
# âž¤ Can handle infinite or large sequences

# Normal list consumes memory:
nums = [i for i in range(1000000)]

# Generator saves memory:
def gen():
    for i in range(1000000):
        yield i

# 8. Lambda function
# âž¤ Anonymous, one-line function. Often used with map(), filter() etc.

square = lambda x: x**2
print(square(4))  # Output: 16

# 8. Lambda function
# âž¤ Anonymous, one-line function. Often used with map(), filter() etc.

square = lambda x: x**2
print(square(4))  # Output: 16

# 9. map() function
# âž¤ Applies a function to all items in an iterable.

nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)  # [1, 4, 9, 16]


# 10. Difference between map(), filter(), reduce()
from functools import reduce

nums = [1, 2, 3, 4, 5]

# map(): transform all values
mapped = list(map(lambda x: x*2, nums))

# filter(): select values
filtered = list(filter(lambda x: x % 2 == 0, nums))

# reduce(): combine values
reduced = reduce(lambda x, y: x + y, nums)

print("Map:", mapped)
print("Filter:", filtered)
print("Reduce:", reduced)

```
