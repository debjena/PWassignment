```python


# âœ… 1. What is hypothesis testing in statistics?
# Hypothesis testing is a statistical method used to make decisions or inferences about population parameters using sample data.

# âœ… 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
# Null Hypothesis (H0): Assumes no effect or difference.
# Alternative Hypothesis (H1): Assumes a significant effect or difference.
# Example: H0: Î¼ = 50, H1: Î¼ â‰  50

# âœ… 3. What is the significance level in hypothesis testing, and why is it important?
# The significance level (Î±) is the probability of rejecting the null hypothesis when it is actually true.
# Common levels: 0.05 or 5%
# It helps control the Type I error rate.

# âœ… 4. What does a P-value represent in hypothesis testing?
# The P-value indicates the probability of observing a result at least as extreme as the sample result, assuming the null hypothesis is true.

# âœ… 5. How do you interpret the P-value in hypothesis testing?
# - If P-value â‰¤ Î±: Reject H0 (result is significant)
# - If P-value > Î±: Fail to reject H0 (result not significant)

# âœ… 6. What are Type 1 and Type 2 errors in hypothesis testing?
# Type I Error (Î±): Rejecting H0 when it's true.
# Type II Error (Î²): Failing to reject H0 when it's false.

# âœ… 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
# - One-tailed: Tests for effect in one direction (e.g., greater than)
# - Two-tailed: Tests for any difference (e.g., â‰ )

# âœ… 8. What is the Z-test, and when is it used in hypothesis testing?
# A Z-test is used to test hypotheses about population means when the population variance is known and the sample size is large (n > 30).

# âœ… 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
# Z = (XÌ„ - Î¼) / (Ïƒ / âˆšn)
# It represents how many standard errors the sample mean is from the population mean.

# âœ… 10. What is the T-distribution, and when should it be used instead of the normal distribution?
# The T-distribution is used when sample size is small (n < 30) and population standard deviation is unknown.

# âœ… 11. What is the difference between a Z-test and a T-test?
# - Z-test: Known population standard deviation, large samples.
# - T-test: Unknown population standard deviation, small samples.

# âœ… 12. What is the T-test, and how is it used in hypothesis testing?
# A T-test is used to compare sample means (one-sample, two-sample, or paired sample) when population standard deviation is unknown.

# âœ… 13. What is the relationship between Z-test and T-test in hypothesis testing?
# Both test hypotheses about means. Z-test is for known Ïƒ, T-test is for unknown Ïƒ.
# As sample size increases, T-distribution approaches normal distribution.

# âœ… 14. What is a confidence interval, and how is it used to interpret statistical results?
# A confidence interval is a range of values that likely contains the population parameter with a given confidence level (e.g., 95%).

# âœ… 15. What is the margin of error, and how does it affect the confidence interval?
# Margin of error is the amount added/subtracted from the point estimate in a confidence interval.
# Larger margin â†’ wider interval â†’ less precision.

# âœ… 16. How is Bayes' Theorem used in statistics, and what is its significance?
# Bayesâ€™ Theorem calculates the probability of an event based on prior knowledge.
# P(A|B) = [P(B|A) * P(A)] / P(B)
# It is used in decision-making, machine learning, medical diagnostics, etc.

# âœ… 17. What is the Chi-square distribution, and when is it used?
# Itâ€™s a distribution used to test categorical data for independence or goodness of fit.
# Itâ€™s skewed right and only takes positive values.

# âœ… 18. What is the Chi-square goodness of fit test, and how is it applied?
# It tests if observed frequencies match expected frequencies.
# Applied by calculating:
# Ï‡Â² = Î£[(O - E)Â² / E], where O = observed, E = expected.

# âœ… 19. What is the F-distribution, and when is it used in hypothesis testing?
# The F-distribution compares two variances.
# Itâ€™s used in ANOVA and regression analysis.

# âœ… 20. What is an ANOVA test, and what are its assumptions?
# ANOVA (Analysis of Variance) tests differences between multiple group means.
# Assumptions:
# - Normality
# - Independence
# - Equal variances (homogeneity)

# âœ… 21. What are the different types of ANOVA tests?
# - One-way ANOVA: One independent variable
# - Two-way ANOVA: Two independent variables
# - Repeated Measures ANOVA: Same subjects under different conditions

# âœ… 22. What is the F-test, and how does it relate to hypothesis testing?
# The F-test compares two population variances.
# Itâ€™s used as the test statistic in ANOVA.

```

```python
# ðŸ“˜ PW Skills - Practical Hypothesis Testing with Python

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# âœ… 1. Z-test: Compare sample mean to known population mean
def z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    print(f"Z-statistic: {z:.4f}, P-value: {p_value:.4f}")
sample = np.random.normal(52, 10, 50)
z_test(sample, 50, 10)

# âœ… 2. Simulate data and perform hypothesis testing with P-value
data = np.random.normal(100, 15, 30)
t_stat, p_val = stats.ttest_1samp(data, 95)
print("One-sample T-test:", "T-statistic =", t_stat, "P-value =", p_val)

# âœ… 3. One-sample Z-test
def one_sample_ztest(sample, pop_mean, pop_std):
    z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    print("Z =", z, "P =", p)
one_sample_ztest(np.random.normal(48, 10, 40), 50, 10)

# âœ… 4. Two-tailed Z-test with decision region visualization
sample = np.random.normal(52, 10, 30)
z_stat = (np.mean(sample) - 50) / (10 / np.sqrt(len(sample)))
alpha = 0.05
z_crit = stats.norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label='Standard Normal')
plt.axvline(z_crit, color='red', linestyle='--', label='Critical Value')
plt.axvline(-z_crit, color='red', linestyle='--')
plt.axvline(z_stat, color='blue', linestyle='-', label='Z-statistic')
plt.legend()
plt.title("Two-tailed Z-test Decision Region")
plt.grid(True)
plt.show()

# âœ… 5. Visualize Type 1 and Type 2 Errors
def visualize_errors(mu0=50, mu1=52, sigma=10, n=30, alpha=0.05):
    x = np.linspace(40, 60, 1000)
    z_crit = stats.norm.ppf(1 - alpha)
    threshold = mu0 + z_crit * (sigma / np.sqrt(n))
    y0 = stats.norm.pdf(x, mu0, sigma / np.sqrt(n))
    y1 = stats.norm.pdf(x, mu1, sigma / np.sqrt(n))

    plt.plot(x, y0, label="H0: Î¼=50", color='blue')
    plt.plot(x, y1, label="H1: Î¼=52", color='green')
    plt.axvline(threshold, color='red', linestyle='--', label="Critical Value")
    plt.fill_between(x, y0, where=(x > threshold), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, y1, where=(x < threshold), color='orange', alpha=0.3, label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.grid(True)
    plt.show()
visualize_errors()

# âœ… 6. Independent T-test
group1 = np.random.normal(55, 5, 30)
group2 = np.random.normal(60, 5, 30)
t_stat, p_val = stats.ttest_ind(group1, group2)
print("Independent T-test:\nT-stat =", t_stat, "P-value =", p_val)

# âœ… 7. Paired Sample T-test & Visualization
before = np.random.normal(50, 5, 20)
after = before + np.random.normal(1.5, 2, 20)
t_stat, p_val = stats.ttest_rel(before, after)
print("Paired T-test:\nT-stat =", t_stat, "P-value =", p_val)
plt.plot(before, label='Before')
plt.plot(after, label='After')
plt.title("Before vs After Comparison")
plt.legend()
plt.grid(True)
plt.show()

# âœ… 8. Compare Z-test and T-test
sample = np.random.normal(100, 15, 30)
pop_mean = 95
z = (np.mean(sample) - pop_mean) / (15 / np.sqrt(len(sample)))
t_stat, p_val = stats.ttest_1samp(sample, pop_mean)
print("Z-test:", z, "\nT-test:", t_stat)

# âœ… 9. Confidence Interval Function
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    se = stats.sem(sample)
    margin = stats.t.ppf((1 + confidence)/2, df=len(sample)-1) * se
    ci = (mean - margin, mean + margin)
    print(f"{int(confidence*100)}% Confidence Interval: {ci}")
confidence_interval(np.random.normal(50, 5, 40))

# âœ… 10. Margin of Error Calculator
def margin_of_error(sample, confidence=0.95):
    se = stats.sem(sample)
    critical = stats.t.ppf((1 + confidence)/2, df=len(sample)-1)
    return critical * se
sample = np.random.normal(50, 5, 30)
print("Margin of Error:", margin_of_error(sample))

# âœ… 11. Bayesian Inference with Bayes' Theorem
# P(A|B) = (P(B|A) * P(A)) / P(B)
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    return (p_b_given_a * p_a) / p_b
result = bayes_theorem(0.01, 0.99, 0.05)
print("Bayesian Inference Result (P(A|B)):", result)

# âœ… 12. Chi-square test for independence
data = pd.DataFrame({'Gender': ['M', 'M', 'F', 'F'], 'Preference': ['Yes', 'No', 'Yes', 'No'], 'Count': [20, 10, 30, 40]})
contingency = data.pivot(index='Gender', columns='Preference', values='Count')
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print("Chi-Square Test:\nChi2 =", chi2, "P =", p)

# âœ… 13. Calculate Expected Frequencies
print("Expected Frequencies:\n", expected)

# âœ… 14. Goodness-of-fit test
observed = [30, 50, 20]
expected = [33.33, 33.33, 33.33]

# Scale expected to match sum of observed exactly
expected = np.array(expected)
expected = expected * (np.sum(observed) / np.sum(expected))

chi2, p = stats.chisquare(f_obs=observed, f_exp=expected)
print("Goodness-of-Fit Test:\nChi2 =", chi2, "P-value =", p)
```
