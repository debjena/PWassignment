```python
# Object-Oriented Programming (OOP) Assignment ‚Äî Answers

---

### 1Ô∏è‚É£ What is Object-Oriented Programming (OOP)?
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of **objects**, which can contain data (attributes) and code (methods). It focuses on organizing code using real-world models, making it reusable, modular, and easier to maintain.

---

### 2Ô∏è‚É£ What is a class in OOP?
A class is a **blueprint or template** for creating objects. It defines the attributes (variables) and behaviors (methods) that the created objects will have.

---

### 3Ô∏è‚É£ What is an object in OOP?
An object is an **instance** of a class. It represents a specific entity that contains the data and behavior defined by its class.

---

### 4Ô∏è‚É£ What is the difference between abstraction and encapsulation?
- **Abstraction** hides complex implementation details and shows only essential features.
- **Encapsulation** wraps data and methods into a single unit (class) and restricts direct access, protecting internal state.

---

### 5Ô∏è‚É£ What are dunder methods in Python?
Dunder (double underscore) methods are **special methods** with names like `__init__`, `__str__`, `__repr__`. They enable operator overloading, object initialization, and control of built-in behaviors.

---

### 6Ô∏è‚É£ Explain the concept of inheritance in OOP.
Inheritance allows a class (child) to **inherit** properties and methods from another class (parent), promoting code reuse and extending functionality.

---

### 7Ô∏è‚É£ What is polymorphism in OOP?
Polymorphism means **one interface, many implementations**. It allows different classes to be treated as the same type through shared methods or interfaces.

---

### 8Ô∏è‚É£ How is encapsulation achieved in Python?
Encapsulation is achieved using **private (`_` or `__`) attributes** and providing public methods (getters/setters) to control access to the data.

---

### 9Ô∏è‚É£ What is a constructor in Python?
A constructor is a special method (`__init__`) that is **automatically called** when a new object is created, initializing the object‚Äôs attributes.

---

### üîü What are class and static methods in Python?
- **Class method** (`@classmethod`) takes `cls` as the first parameter and can modify class state.
- **Static method** (`@staticmethod`) does not take `self` or `cls` and behaves like a regular function inside the class.

---

### 11Ô∏è‚É£ What is method overloading in Python?
Method overloading means **defining multiple methods with the same name** but different parameters. Python doesn‚Äôt support it directly but can mimic it using default or variable arguments.

---

### 12Ô∏è‚É£ What is method overriding in OOP?
Method overriding means redefining a **parent class method** in a child class to change or extend its behavior.

---

### 13Ô∏è‚É£ What is a property decorator in Python?
The `@property` decorator allows a method to be accessed like an **attribute** (without parentheses), often used for computed or protected values.

---

### 14Ô∏è‚É£ Why is polymorphism important in OOP?
Polymorphism provides **flexibility and scalability** by allowing code to handle objects of different classes through a common interface.

---

### 15Ô∏è‚É£ What is an abstract class in Python?
An abstract class is a class with **abstract methods** (declared but not implemented) and is used to define a common interface for its subclasses.

---

### 16Ô∏è‚É£ What are the advantages of OOP?
- Code reuse through inheritance
- Better organization and modularity
- Easier maintenance
- Encapsulation for data protection
- Abstraction to manage complexity

---

### 17Ô∏è‚É£ What is the difference between a class variable and an instance variable?
- **Class variable** is shared across all instances.
- **Instance variable** is unique to each object.

---

### 18Ô∏è‚É£ What is multiple inheritance in Python?
Multiple inheritance allows a class to **inherit from multiple parent classes**, combining behaviors from all.

---

### 19Ô∏è‚É£ Explain the purpose of `__str__` and `__repr__` methods in Python.
- `__str__` defines the **human-readable** string representation (`str()`).
- `__repr__` defines the **official or developer-facing** string (`repr()`).

---

### 20Ô∏è‚É£ What is the significance of the `super()` function in Python?
`super()` allows calling **methods from the parent class**, enabling proper initialization and method extension in inheritance chains.

---

### 21Ô∏è‚É£ What is the significance of the `__del__` method in Python?
`__del__` is the **destructor** method, called when an object is about to be destroyed, used for cleanup operations.

---

### 22Ô∏è‚É£ What is the difference between `@staticmethod` and `@classmethod` in Python?
- `@staticmethod` has no access to class or instance.
- `@classmethod` receives the class (`cls`) and can access/modify class-level data.

---

### 23Ô∏è‚É£ How does polymorphism work in Python with inheritance?
Polymorphism works by allowing child classes to **override parent methods** and still be accessed via parent references, enabling dynamic behavior.

---

### 24Ô∏è‚É£ What is method chaining in Python OOP?
Method chaining is calling **multiple methods in a single line** by having each method return `self`.

---

### 25Ô∏è‚É£ What is the purpose of the `__call__` method in Python?
The `__call__` method makes an object **callable like a function**, allowing it to be invoked with parentheses.

```

```python
# Practical Question 1
class Animal:
    def speak(self):
        print("Animal speaks something")

class Dog(Animal):
    def speak(self):
        print("Bark!")

d = Dog()
d.speak()

# Practical Question 2
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

c = Circle(5)
r = Rectangle(4, 6)
print(c.area())
print(r.area())

# Practical Question 3
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

e_car = ElectricCar("Car", "Tesla", "100 kWh")
print(e_car.type, e_car.brand, e_car.battery)

# Practical Question 4
class Bird:
    def fly(self):
        print("Bird is flying")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies quickly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

s = Sparrow()
p = Penguin()
s.fly()
p.fly()

# Practical Question 5
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")

    def check_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.check_balance())

# Practical Question 6
class Instrument:
    def play(self):
        print("Instrument is playing")

class Guitar(Instrument):
    def play(self):
        print("Guitar is strumming")

class Piano(Instrument):
    def play(self):
        print("Piano is playing")

g = Guitar()
p = Piano()
g.play()
p.play()

# Practical Question 7
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(5, 3))
print(MathOperations.subtract_numbers(5, 3))

# Practical Question 8
class Person:
    count = 0

    def __init__(self):
        Person.count += 1

p1 = Person()
p2 = Person()
p3 = Person()
print("Total persons:", Person.count)

# Practical Question 9
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

f = Fraction(3, 4)
print(f)

# Practical Question 10
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3)

# Practical Question 11
class PersonWithDetails:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

p = PersonWithDetails("Alice", 30)
p.greet()

# Practical Question 12
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

s = Student("John", [85, 90, 78])
print(s.average_grade())

# Practical Question 13
class Rectangle:
    def set_dimensions(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

r = Rectangle()
r.set_dimensions(4, 5)
print(r.area())

# Practical Question 14
class Employee:
    def __init__(self, hours, rate):
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate

class Manager(Employee):
    def __init__(self, hours, rate, bonus):
        super().__init__(hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

e = Employee(40, 20)
m = Manager(40, 20, 200)
print(e.calculate_salary())
print(m.calculate_salary())

# Practical Question 15
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

p = Product("Laptop", 1000, 2)
print(p.total_price())

# Practical Question 16
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"

c = Cow()
s = Sheep()
print(c.sound())
print(s.sound())

# Practical Question 17
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

b = Book("1984", "George Orwell", 1949)
print(b.get_book_info())

# Practical Question 18
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

m = Mansion("123 Luxury St", 5000000, 10)
print(m.address, m.price, m.number_of_rooms)
```
